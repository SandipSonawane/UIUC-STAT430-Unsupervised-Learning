## 3. Gaussian Mixture Models Parameter Selection

We would like to use Gaussian Mixture models to cluster this dataset. However, we are currently unsure as to how many clusters we should ask for.

### 3.1. Parsimonious Gaussian Mixture Model

We would first like to assess the number of clusters $k$ which will yield a parsimonious Gaussian mixture model for this dataset. 

#### 3.1.1. Reliability of AIC for Selecting a Parsimonious Model

For what values of the cluster number $k$ can we rely on the AIC score to produce useful insights regarding the the parsimonious model? In general, can we rely on the AIC score at all with this dataset?



> The rule of thumb to use AIC that, we should use AIC when below condition is true:

$$ \frac{m}{p} > 40  $$

    m: Number of objects in data
    p: Number of parameters

    - We can see there are about 12 clusters in the t-SNE plot, hence there will be 84 parameters to be estimated.
    - Since ratio of m and p is less than 40, AIC won't be a reliable for producing useful insights regarding the parsimonious model

```python
p = 12*(1+784+784**2)
p
```

```python
m = 500
m/p
```

```python
AIC_good = []
AIC_not_good = []
for k in np.arange(1, 12, 1):
    p = k*(1+784+784**2)
    if m/p > 40:
        AIC_good.append(k)
    else:
        AIC_not_good.append(k)
```

```python
AIC_good, AIC_not_good
```

#### 3.1.2. AIC and BIC Score Plots

Create two plots.
1. One plot should plot the AIC score of a Gaussian mixture model clustering with k=1, k=2, ..., k=20 clusters respectively.
2. The other plot should plot the BIC score of a Gaussian mixture model clustering with k=1, k=2, ..., k=20 clusters respectively.

Use a random state of 101 for all of these clusterings.

```python
aics = []
bics = []
log_likelihoods = []

for k in range(1, 20):
    #Fit the GMM model
    gm = GaussianMixture(n_components=k, random_state=101)
    gm.fit(X)
    
    #Calculate the AIC, BIC, and log-likelihood for this model
    aics.append(gm.aic(X))
    bics.append(gm.bic(X))
    log_likelihoods.append(gm.score(X) * len(X))


#Plot these values
fig, ax = plt.subplots(1, 3, figsize=(30, 6))

#AIC
ax[0].plot(range(1, 20), aics)
ax[0].set_xticks(range(1, 20))
ax[0].set_xlabel('k', fontsize=16)
ax[0].set_title('AIC', fontsize=16)

#BIC
ax[1].plot(range(1, 20), bics)
ax[1].set_xticks(range(1, 20))
ax[1].set_xlabel('k', fontsize=16)
ax[1].set_title('BIC', fontsize=16)

#Log-likelihood
ax[2].plot(range(1, 20), log_likelihoods)
ax[2].set_xticks(range(1, 20))
ax[2].set_xlabel('k', fontsize=16)
ax[2].set_title('Log-likelihood', fontsize=16)

plt.show()
```

#### 3.1.3 Interpretation

Use the AIC score to select the number of clusters that will yield a parsimonious Gaussian mixture model. Use the BIC score to select the number of clusters that will yield a parsimonious Gaussian mixture model.

> The BIC score is suggesting that a Gaussian mixture model clustering with k=1 clusters will yield a parsimonious model, while the AIC score is suggesting that a Gaussian mixture model clustering with k=2 clusters will yield a parsimonious model. Let's try out both cluster numbers.

#### 3.1.4. Usefulness

Given that our goal is to determine the underlying clusters of this dataset, particularly those that the t-SNE plot is suggesting might exist, do we think that the cluster number $k$ suggested by the AIC or BIC score are useful?

> The clusters suggested by AIC as well as BIC are not useful since it is recommending far less clusters than the actual number of clusters in the t-SNE plot.

### 3.2. t-SNE Plot and GMM with k=10 clusters

#### 3.2.1. Cluster the Data
Cluster the dataset into k=10 clusters using Gaussian mixture models. Use a random state of 101.

```python
#Instantiate the GMM object gm (with your desired algorithm parameters)
gm = GaussianMixture(n_components=10, random_state=101)

#Run the GMM-EM algorithm on your data
gm.fit(X)

#Get the cluster membership scores (ie. posterior probabilities, P(x in Cluster_k| x)))
W=gm.predict_proba(X)
print(np.shape(W))
W
```

```python
cluster_columns=['Cluster %s Membership Score'%(str(i)) for i in range(0,10)]
df_membership=pd.DataFrame(W, columns=cluster_columns)
df_membership.round(4)
```

#### 3.2.2. t-SNE Plots

Then, plot k=10 t-SNE plots. In the kth plot, you should color-code the objects by their kth cluster membership weight (ie. their kth cluster posterior value).

```python
num_row = 2
num_col = 5

# get a segment of the dataset
num = num_row*num_col
labels = df_membership.columns

# plot images
fig, axes = plt.subplots(num_row, num_col, figsize=(4*num_col,3.5*num_row))

for i in range(num_row* num_col):
    #Create a column with cluster i membership scores
    df_combo['Cluster %s Membership Score'%str(i)]=df_membership['Cluster %s Membership Score'%(str(i))]

    #Plot the points, color coded by the membership score.
    points = plt.scatter(df_combo["x_projected"], df_combo["y_projected"],
                 c=df_combo['Cluster %s Membership Score'%str(i)],
                         s=30, cmap="PuRd",
                        vmin=0,
                        vmax=1)
    ax = axes[i//num_col, i%num_col]
    ax.scatter(df_combo["x_projected"], df_combo["y_projected"],
                 c=df_combo['Cluster %s Membership Score'%str(i)],
                                            s=30, cmap="PuRd",
                                           vmin=0,
                                           vmax=1)

    fig.colorbar(points, ax=ax)
    ax.set_title('Cluster %s Membership Scores'%(str(i)))

plt.tight_layout()
plt.show()
```

#### 3.2.3. Visualize the Cluster Means

Finally, visualize the 10 cluster mean images below.

```python
labels_GMM = gm.predict(X)
labels_GMM[:10]
```

```python
df_gmm_averages = X.copy()
df_gmm_averages['lable_gmm'] = labels_GMM
df_gmm_averages=df_gmm_averages.groupby(['lable_gmm']).mean()
df_gmm_averages
```

```python
# specify the number of rows and columns you want to see
num_row = 2
num_col = 5

# get a segment of the dataset
num = num_row*num_col
labels = df_gmm_averages.index[:num]

# plot images
fig, axes = plt.subplots(num_row, num_col, figsize=(2.5*num_col,3.5*num_row))
for i in range(num_row*num_col):
    img_obj = np.array(df_gmm_averages.iloc[i,:]).reshape((28,28))*255
    ax = axes[i//num_col, i%num_col]
    ax.imshow(img_obj, cmap='gray')
    ax.set_title('Label: {}'.format(labels[i]))
    
plt.tight_layout()
plt.show()
```

For each of these images, we will assume that they were generated using the same random process that created our original dataset. Display each of the resulting images. 

We will set the random seed of 101 below. Also, when generating each of the 6 points from their respective Gaussian distribution, you should set a random state of 101 as well within the corresponding random variable value generating function.

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.rv_continuous.rvs.html

First we randomly select which of the 12 Gaussian distributions (ie. clusters) that we would like to randomly generate a data point from.

### 6. One Step of the EM Algorithm

We would like to cluster the data in the **artificial_small.csv** using Gaussian Mixture Model clustering using k=2 clusters (ie. Gaussian distributions).  However, we will only perform one iteration of the EM algorithm and we will do it "by hand". This dataset has only 6 objects. 

Suppose the EM algorithm has randomly initialized the following parameters.

**Cluster 0**:
* $\mu_0 = [1.3, 0.6]$
* $\Sigma_0 = [[0.25, 0],[0,0.25]]$
* $P(x\in C_0)=0.4$

**Cluster 1**:
* $\mu_1 = [1.8, 1.1]$
* $\Sigma_1 = [[0.5, 0],[0,0.5]]$
* $P(x\in C_1)=0.6$

Determine what the new **means** and **priors** will be in the next iteration of the algorithm for both clusters. Show your work. You can use Python functions and code for your calculations.

```python
df_6 = pd.read_csv('artificial_small.csv')
mu_0=[1.3,0.6]
mu_1 = [1.8,1.1]

plt.scatter(df_6.x1, df_6.x2)
plt.scatter(mu_0[0], mu_0[1], c='r', label='cluster 0 mean')
plt.scatter(mu_1[0], mu_1[1], c='black', label='cluster 1 mean')
plt.legend()
plt.show()
```

### Initialization Step

```python
mu_0=[1.3,0.6]
sigma_0=[[0.25,0],[0,0.25]]
prior_c0 = 0.4

mu_1 = [1.8,1.1]
sigma_1 = [[0.5,0],[0,0.5]]
prior_c1 = 0.6
```

### Expectation Step


```python
def calc_expectation(obj_idx, mu, sigma, prior):
    
    numerator = multivariate_normal.pdf(df_6.iloc[obj_idx,:], mu, sigma)*prior
    
    denominator = multivariate_normal.pdf(df_6.iloc[obj_idx,:], mu_0, sigma_0)*prior_c0 + \
        multivariate_normal.pdf(df_6.iloc[obj_idx,:], mu_1, sigma_1)*prior_c1
    
    expecation = numerator/denominator
    
    return expecation
```

```python
df_posterior = pd.DataFrame(columns=['Cluster 0 posterior', 'Cluster 1 posterior'])
```

```python
for i in range(len(df_6)):
    cls_0 = calc_expectation(i, mu_0, sigma_0, prior_c0)
    cls_1 = calc_expectation(i, mu_1, sigma_1, prior_c1)
    df_posterior = df_posterior.append({'Cluster 0 posterior':cls_0, 'Cluster 1 posterior':cls_1}, ignore_index=True)
```
