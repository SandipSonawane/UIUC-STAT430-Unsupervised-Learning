## 3. NMF Parameter Selection

### 3.1. Elbow Plot

Construct a NMF elbow plot (see case study 1 in this assignment #5 for how we did this). This plot should plot the optimal objective function value for 50 NMF clusterings, where each clustering has k = 1, 2, ..., 50 clusters. Each clustering should be created with a random state of 100.

```python
nmf_reconstruction_error = []
for k in range(1, 50):
    nmf_k = NMF(n_components=k, random_state=100)
    nmf_k.fit_transform(X)
    nmf_reconstruction_error.append(nmf_k.reconstruction_err_)
```

```python
plt.plot(np.arange(1,50,1), nmf_reconstruction_error)
plt.stem(np.arange(1,50,1), nmf_reconstruction_error, '--')
plt.xticks(np.arange(1,50,3))
plt.ylim([30, 160])
plt.title('Elbow Plot NMF')
plt.xlabel('No. of clusters')
plt.ylabel('NMF reconstruction error')
plt.show()
```

## 4. NMF Clustering

Let's first use $k=10$ clusters because this corresponds to about how many clusters are suggested by the t-SNE plot (of the images) and we have 10 image classes.

### 4.1. NMF Clustering 

Cluster your dataset using NMF with k=10 clusters. Use a random state of 100. If your dataset is $X $and you approximate $X\approx W\cdot H$ in your decomposition, display your W and H matrices in two dataframes.

```python
# Perform a Non-negative matrix factorization
nmf = NMF(n_components=10, random_state=100)
W=nmf.fit_transform(X)
H=nmf.components_
```

```python
df_W=pd.DataFrame(W, columns=['cluster_'+str(k) for k in range(0,10)])
df_W.round(4)
```

```python
df_H = pd.DataFrame(H)
df_H
```

### 4.2. Display Results

Plot your t-SNE plot ten times: each time color coding the points by one of the columns in W.

```python
num_row = 5
num_col = 2

# get a segment of the dataset
num = num_row*num_col

# plot images
fig, axes = plt.subplots(num_row, num_col, figsize=(4*num_col,3.5*num_row))

perp, rs = 30, 103
tsne = TSNE(n_components=2, perplexity=perp, random_state=rs)
data_tsne = tsne.fit_transform(X)
df_tsne = pd.DataFrame(data_tsne, columns=['x_projected', 'y_projected'])
df_combo = pd.concat([df, df_tsne], axis=1)
    
for i in range(num_row* num_col):
    ax = axes[i//num_col, i%num_col]
    
    
    sns.scatterplot(x='x_projected',y='y_projected', hue=df_W.iloc[:, i], data=df_combo, ax=ax)
#     plt.legend(bbox_to_anchor=(1,1))
    ax.set_title(f't-SNE with w{i}')

plt.tight_layout()
plt.show()
```

### 4.3. Image Cluster Membership Matrix

Create a cluster membership matrix for for the images, using the W matrix.

```python
cm_scores_W=W/np.sum(W, axis=1, keepdims=True)
cm_scores_W[:5]
```

```python
df_cm_scores_W=pd.DataFrame(cm_scores_W, columns=df_W.columns)
df_cm_scores_W
```

### 4.4. Interpretation

To what extent does the 13th image in the dataset belong to each of the 10 image clusters?

```python
plt.figure(figsize=(8,8))
df_cm_scores_W.iloc[12,:].plot.pie(autopct='%1.1f%%')
plt.show()
```

> We can see the object belongs to cluster 7 with 51.7% cluster membership score. It belongs to cluster 9 with 24.6% cluster membership score. For other clusters, the cluster membership scores are low.

### 4.5. Displaying the Cluster Membership Values

Plot your t-SNE plot ten times: each time color coding the points by the ten cluster membership score values.

```python
num_row = 5
num_col = 2

# get a segment of the dataset
num = num_row*num_col

# plot images
fig, axes = plt.subplots(num_row, num_col, figsize=(4*num_col,3.5*num_row))

perp, rs = 30, 103
tsne = TSNE(n_components=2, perplexity=perp, random_state=rs)
data_tsne = tsne.fit_transform(X)
df_tsne = pd.DataFrame(data_tsne, columns=['x_projected', 'y_projected'])
df_combo = pd.concat([df, df_tsne], axis=1)
    
for i in range(num_row* num_col):
    ax = axes[i//num_col, i%num_col]
    
    
    sns.scatterplot(x='x_projected',y='y_projected', hue=df_cm_scores_W.iloc[:, i], data=df_combo, ax=ax)
#     plt.legend(bbox_to_anchor=(1,1))
    ax.set_title(f't-SNE with w_cm_{i}')

plt.tight_layout()
plt.show()
```

### 4.6 Strong Cluster Membership Scores

Create a histogram which displays the highest cluster membership score for each of the 500 images.

```python
cm_max_scores = df_cm_scores_W.max(axis=1)
plt.hist(cm_max_scores, bins=25)
plt.show()
```

> There are more number of objects having max cluster membership score around 0.3 to 0.4. This tells us that its a soft clustering.
